```python
# File mergedexc.py (Python 3.X + 2.X)
sep = '-' * 45 + '\n' # ---------------------------------------------

print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x = 'spam'[99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

##### **`EXCEPTION RAISED AND CAUGHT`**
- `except run` :

    `x = 'spam'[99]` es un `IndexError`, porque `'spam'[99]` esta fuera de el largo de el string `'spam'` que es 4

```python
print(sep + 'NO EXCEPTION RAISED')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

##### **`NO EXCEPTION RAISED`**
`x = 'spam'[3]` si que existe: (`'m'`)

Por lo tanto no salta la excepción

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print('else run')
finally:
    print('finally run')
print('after run')
```

##### **`NO EXCEPTION RAISED, WITH ELSE`**

- `else run` :

    el `else` en el try-except se usa para hacer algo cuando **NO** ocurrió ningun error

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x = 1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

##### **`EXCEPTION RAISED BUT NOT CAUGHT`**

no encuentra la excepción porque `x = 1 / 0` no es un `IndexError` si no un `ZeroDivisionError`
