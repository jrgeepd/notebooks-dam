## üß† Conceptos clave
Un **m√≥dulo** es un archivo Python (`.py`) que contiene c√≥digo reutilizable (funciones, clases o variables).
Los m√≥dulos permiten **organizar y reutilizar c√≥digo** en programas m√°s grandes.


## üß© Creaci√≥n de un m√≥dulo
Un m√≥dulo se crea simplemente guardando c√≥digo en un archivo `.py`.

```python
# archivo: hola.py
def saludar(nombre):
    print(f"Hola, {nombre}!")
```


````
# Ejemplo: importaci√≥n b√°sica
import hola
hola.saludar('Pythonista')
````

## üîß Formas de importar m√≥dulos
- `import modulo`: importa todo el m√≥dulo.
- `from modulo import nombre`: importa un nombre espec√≠fico.
- `import modulo as alias`: importa con un alias.
- `from modulo import *`: importa todo (‚ö†Ô∏è no recomendado).


````
# Ejemplos de importaciones
from hola import saludar
saludar('Mundo')

import hola as h
h.saludar('Alumno')
````

## ‚ôªÔ∏è Recarga de m√≥dulos
Cuando el c√≥digo de un m√≥dulo cambia, Python **no lo recarga autom√°ticamente**.
Para recargarlo, usa `importlib.reload()`.


````
from importlib import reload
reload(hola)
````

## üß≠ Espacios de nombres de los m√≥dulos
Cada m√≥dulo tiene su propio **espacio de nombres**, que es un diccionario con sus variables y funciones.

Puedes explorar este espacio con `dir()` o accediendo al atributo `__dict__` del m√≥dulo.

````
print(dir(hola))
print(hola.__dict__.keys())
````

## üìÇ Estructura de archivos y m√≥dulos
- Cada archivo `.py` crea un espacio de nombres independiente.
- Los nombres se acceden como atributos del m√≥dulo.

```python
import math
print(math.pi)
```


## üß∞ Buenas pr√°cticas
- Evita `from modulo import *`.
- Usa `if __name__ == '__main__':` para c√≥digo de prueba.
- Usa `reload()` solo durante el desarrollo.
- Mant√©n los nombres claros y consistentes.
